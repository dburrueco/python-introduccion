```python
intentos = 0
right_password = "abc"

while intentos < 3:
    password = input("Introduce la contraseña")
    if password == right_password:
        print("Acceso concedido")
        break
    intentos += 1
else:
    print("Has agotado el número máximo de intentos. Tu cuenta se ha bloqueado")
```
