### break

```python
numbers = [4, 1, 8, -2, 3, -1]
```

```python
for number in numbers:
    print(number)
```

```python
for number in numbers:
    if number < 0:
        break
    print(number)
```

### else

```python
for number in numbers:
    if number == 99:
        print("Número encontrado")
        break
else:
    print("El valor no ha sido encontrado")
```

```python
for number in numbers:
    if number < 0:
        continue
    print(number)
```

### enumerate

```python
clientes = ["Juan", "Beatriz", "Luis"]
```

```python
contador = 0
for cliente in clientes:
    print(contador, cliente)
    contador += 1
```

```python
for contador, cliente in enumerate(clientes, start = 1):
    print(contador, cliente)
```
